```python
import pandas as pd
import os
import shutil
import random

pd.set_option('display.max_columns',None)
```

```python
# creating data for positive 

FILE_PATH = 'covid_positive/metadata.csv'
IMAGES_PATH = 'covid_positive/images/'
TARGET_PATH = 'Dataset/Covid'
```

```python
if not os.path.exists(TARGET_PATH):
    os.makedirs(TARGET_PATH)
```

```python
positive_data = pd.read_csv(FILE_PATH)
positive_data.head()
```

```python
# we want only that data that has COVID-19 in findings column and PA in view column(PA means Front View of X ray), we 
# don't want Top or any other view of X ray

# count of covid 19 and Front view X rays
count = 0
for i in positive_data.itertuples():
    if 'COVID-19' in i.finding  and i.view == 'PA':
        # moving image from covid positive to new Dataset folder 
        filename = i.filename
        image_path = os.path.join(IMAGES_PATH,filename)
        image_copy_path = os.path.join(TARGET_PATH,filename)
        shutil.copy2(image_path,image_copy_path)
        print("Moving",count)
        count += 1
```

```python
count
```

```python
## Getting Normal Images 
# Normal Images are in kaggle downloaded folder

NORMAL_FILEPATH = 'Pnemonia_data/train/NORMAL/'
TARGET_NORMAL_PATH = 'Dataset/Normal'

if not os.path.exists(TARGET_NORMAL_PATH):
    os.makedirs(TARGET_NORMAL_PATH)
    print('TARGET NORMAL directory created')

# There are alot of images available in Normal folder, we will take 196 Normal Images
normal_filenames = os.listdir(NORMAL_FILEPATH)

# we need random 196 images 
random.shuffle(normal_filenames)

# getting 196 images
for i in range(196):
    filename = normal_filenames[i]
    filepath = os.path.join(NORMAL_FILEPATH,filename)
    target_filepath = os.path.join(TARGET_NORMAL_PATH,filename)
    shutil.copy2(filepath,target_filepath)
    print('moving',i)
```

```python
## Making Train and Test split of data
covid_train_path = 'Dataset/Train/Covid/'
covid_test_path = 'Dataset/Test/Covid/'
normal_train_path = 'Dataset/Train/Normal/'
normal_test_path = 'Dataset/Test/Normal'

if not os.path.exists(covid_train_path):
    os.makedirs(covid_train_path)
    print("covid_train_path directory created")

if not os.path.exists(covid_test_path):
    os.makedirs(covid_test_path)
    print("covid_test_path directory created")
    
if not os.path.exists(normal_train_path):
    os.makedirs(normal_train_path)
    print("normal_train_path directory created")
    
if not os.path.exists(normal_test_path):
    os.makedirs(normal_test_path)
    print("normal_test_path directory created")
```

```python
training_size = int(len(os.listdir(TARGET_NORMAL_PATH))*0.8)
test_size = len(os.listdir(TARGET_NORMAL_PATH)) - training_size
total_size = len(os.listdir(TARGET_NORMAL_PATH))
```

```python
training_size,test_size
```

```python
# getting training data in to train folder
COVID_PATH = 'Dataset/Covid/'
NORMAL_PATH = 'Dataset/Normal/'

covid_files = os.listdir(COVID_PATH)
normal_files = os.listdir(NORMAL_PATH)

random.shuffle(covid_files)
random.shuffle(normal_files)
```

```python
# for covid
for i in range(total_size):
    covid_image_name = covid_files[i]
    image_path = os.path.join(COVID_PATH,covid_image_name)
    if i < training_size:
        target_train_covid_image_path = os.path.join(covid_train_path,covid_image_name)
        shutil.move(image_path,target_train_covid_image_path)
        print("moved ",i)
    else:
        target_test_covid_image_path = os.path.join(covid_test_path,covid_image_name)
        shutil.move(image_path,target_test_covid_image_path)
        print("moved",i)
        
```

```python
# for normal
for i in range(total_size):
    normal_image_name = normal_files[i]
    image_path = os.path.join(NORMAL_PATH,normal_image_name)
    if i < training_size:
        target_train_normal_image_path = os.path.join(normal_train_path,normal_image_name)
        shutil.move(image_path,target_train_normal_image_path)
        print("moved ",i)
    else:
        target_test_normal_image_path = os.path.join(normal_test_path,normal_image_name)
        shutil.move(image_path,target_test_normal_image_path)
        print("moved",i)
```
